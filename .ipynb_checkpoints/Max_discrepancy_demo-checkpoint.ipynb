# 导入模块

```python
import tensorflow as tf
import random
import numpy as np
import spectral as sp
import HSI_data_Class as hc
from cnn_demo import pool
from cnn_demo import batch_norm as BN 
from cnn_demo import fc_batch_norm as FBN
from tensorflow.python.framework import ops
```

# 数据准备

```python
'''生成两种数据，一种采样百分之五作为源数据，一种采样百分之百作为目标数据，size=7，数据格式都为[batch,7,7,200]'''
size=7
data=hc.hsi_data('I')
dpt=data.img.shape[2]
sample1=hc.sample_hsi(data.img_gt,0.05)
sample1.generate_ss(data.img,size)
sample2=hc.sample_hsi(data.img_gt,1.0)
sample2.generate_ss(data.img,size)
class_num=len(np.unique(data.img_gt))
dpt=data.img.shape[2]
```

# 数据读取

```python
tf.reset_default_graph()
#batch=200,前100是源数据，后100是目标数据
x_1=tf.placeholder('float',[None,size,size,dpt])
y_=tf.placeholder('float',[None,17])
dp=tf.placeholder('float')
training=tf.placeholder('bool')
domain_cls=tf.placeholder('float',[None,2])#并没有用，原来的代码改的
```

# 网络结构

```python
'''结构分为三个部分，一个特征提取器，两个分类器，第一阶段正常训练，第二阶段固定特征提取器，调整两个分类器，最小化分类误差的同时，
   增大两个分类器在目标数据上面的差异，第三阶段固定两个分类器，调整特征提取器使得两个分类器在目标数据上面的差异变小（我理解的第二
   阶段是找到源数据和目标数据分布不一致数据的边界，然后第三阶段通过调整特征，将特征调整进边界内，可以结合论文图来理解），我是用三
   个阶段交替来训练的，因为论文说阶段的顺序不重要0.0'''
with tf.variable_scope('feature'):
    con_w0=tf.Variable(tf.random_normal([3,3,200,128]))
    con_b0=tf.Variable(tf.random_normal([128]))
    conv0=tf.nn.relu( BN(tf.nn.conv2d(BN(x_1,training), con_w0, strides=[1, 1, 1, 1], padding='SAME'),training)+con_b0)
    
    conv0=pool(conv0)
    
    con_w1=tf.Variable(tf.random_normal([3,3,128,128]))
    con_b1=tf.Variable(tf.random_normal([128]))
    conv1=tf.nn.relu( BN(tf.nn.conv2d(conv0, con_w1, strides=[1, 1, 1, 1], padding='SAME'),training)+con_b1)
    
    conv1=pool(conv1)
    
    con_w2=tf.Variable(tf.random_normal([3,3,128,128]))
    con_b2=tf.Variable(tf.random_normal([128]))
    conv2=tf.nn.relu( BN(tf.nn.conv2d(conv1, con_w2, strides=[1, 1, 1, 1], padding='SAME'),training)+con_b2)
    
    conv2=tf.reshape(conv2,[-1,2*2,128])
    conv2=tf.reduce_mean(conv2,axis=1)

with tf.variable_scope('class1'):
    w1=tf.Variable(tf.random_normal([128,128]))
    b1=tf.Variable(tf.random_normal([128]))
    fc1=tf.nn.sigmoid(FBN(tf.matmul(conv2,w1)+b1,training))
    
    fc1=tf.nn.dropout(fc1,dp)
    
    w2=tf.Variable(tf.random_normal([128,17]))
    b2=tf.Variable(tf.random_normal([17]))
    fc2_1=tf.matmul(fc1,w2)+b2
    
    output1=tf.nn.softmax(fc2_1)#额外的节点

with tf.variable_scope('class2'):
    w1=tf.Variable(tf.random_normal([128,128]))
    b1=tf.Variable(tf.random_normal([128]))
    fc1=tf.nn.sigmoid(FBN(tf.matmul(conv2,w1)+b1,training))
    
    fc1=tf.nn.dropout(fc1,dp)
    
    w2=tf.Variable(tf.random_normal([128,17]))
    b2=tf.Variable(tf.random_normal([17]))
    fc2_2=tf.matmul(fc1,w2)+b2
    
    output2=tf.nn.softmax(fc2_2)#额外的节点
    
#分类误差损失
loss_cls1=tf.reduce_sum(tf.nn.softmax_cross_entropy_with_logits(labels=y_[:100],logits=fc2_1[:100],dim=1))#前100数据为源数据，用于分类
loss_cls2=tf.reduce_sum(tf.nn.softmax_cross_entropy_with_logits(labels=y_[:100],logits=fc2_2[:100],dim=1))
loss_cls=loss_cls1+loss_cls2
#两个分类器的绝对误差损失
loss_drpc=tf.reduce_sum(tf.abs(output1[100:]-output2[100:]))#后100为目标数据，用于衡量分类差异

'''获得变量域里面的变量，用于不同阶段的训练'''
var_list_1 = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='class1')#分类器一域
var_list_2=tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='class2')#分类器二域
var_list_3 = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='feature')#特征域


train_step1= tf.train.AdamOptimizer(0.005).minimize(loss_cls)#阶段一，全部训练
train_step2= tf.train.AdamOptimizer(0.005).minimize(loss_cls-loss_drpc,var_list=[var_list_2,var_list_1])#阶段二，调整分类器
train_step3= tf.train.AdamOptimizer(0.005).minimize(loss_drpc,var_list=var_list_3)#阶段三，调整特征提取部分
init = tf.global_variables_initializer()
sess = tf.InteractiveSession()
sess.run(init)
```

# 保存和记录

```python
saver = tf.train.Saver()
'''记录节点'''
tf.summary.scalar('loss_cls1', loss_cls1)
tf.summary.scalar('loss_cls2', loss_cls2)
tf.summary.scalar('loss_drpc',loss_drpc)
merged = tf.summary.merge_all()
name='max_descrepancy'
train_writer = tf.summary.FileWriter( 'tf_log/{}'.format(name), sess.graph)
```

# 训练

```python
'''训练部份'''
a=1000
for i in range(6000):
    '''源数据'''
    index_val=random.sample(range(len(sample1.train_feat)),100)
    index_val.sort()
    x1_1=np.array(sample1.train_feat[index_val])
    y_1=np.array(sample1.train_label[index_val])
    '''目标数据'''
    index_val=random.sample(range(len(sample2.train_feat)),100)
    index_val.sort()
    x1_2=np.array(sample2.train_feat[index_val])
    y_2=np.array(sample2.train_label[index_val])
    '''合并，前一百是源数据，后一百是目标数据'''
    x1=np.concatenate([x1_1,x1_2],axis=0)
    y=np.concatenate([y_1,y_2],axis=0)
    '''三个阶段交替训练'''
    if int(i/1) %3==0:
       sess.run(train_step1,feed_dict={x_1:x1,y_:hc.one_hot(y,17),dp:0.5,training:True})
    elif int(i/1) %3==1:
       sess.run(train_step2,feed_dict={x_1:x1,y_:hc.one_hot(y,17),dp:0.5,training:True})
    else:
       sess.run(train_step3,feed_dict={x_1:x1,y_:hc.one_hot(y,17),dp:0.5,training:True})
    '''记录'''
    if i %10==0:
        #验证集用的目标数据，也可以改成源数据
        index_val=random.sample(range(len(sample2.train_feat)),100)
        index_val.sort()
        x1_1=np.array(sample2.train_feat[index_val])
        y_1=np.array(sample2.train_label[index_val])
        
        x1=np.concatenate([x1_1,x1_2],axis=0)
        y=np.concatenate([y_1,y_2],axis=0)
        
        l,l3,_=sess.run([loss_cls,loss_drpc,merged],feed_dict={x_1:x1,y_:hc.one_hot(y,17),dp:1.0,training:False})
        print('iter{},loss_cls:{},loss_drpc:{}'.format(i,l,l3))
        train_writer.add_summary(_,i)
        if l<a:
           a=l
           saver.save(sess,'Model_1231_cnn_max_descrepancy/mnist.ckpt')
```

# 测试


```python
saver.restore(sess,'Model_1231_cnn_max_descrepancy/mnist.ckpt')
sh=data.img.shape
result=np.zeros((sh[0],sh[1]))
for i in range(sh[0]):
    sample2.generate_ss_base_location(data.img,size,[i],range(sh[1]))
    test_feat=np.array(sample2.test_feat)
    te_sen1=test_feat[:,:,:,:]
    #te_sen2=test_feat[:,:,:,15:] 

    x1=np.array(te_sen1).reshape((-1,size,size,dpt))
    #x2=np.array(te_sen2).reshape((-1,9,9,15))
    temp1=sess.run(output1,feed_dict={x_1:x1,training:False,dp:1.0})
    temp2=sess.run(output2,feed_dict={x_1:x1,training:False,dp:1.0})
    result[i]=(temp1+temp2).argmax(axis=1)
result=np.uint8(result)
result[np.where(data.img_gt==0)]=0
sp.imshow(classes=result)
sp.imshow(classes=data.img_gt)
from sklearn.metrics import accuracy_score,classification_report
print(classification_report(data.img_gt.reshape(-1)[np.where(data.img_gt.reshape(-1)!=0)]\
                            ,result.reshape(-1)[np.where(data.img_gt.reshape(-1)!=0)]))
```
